```python
import matplotlib
matplotlib.use('Agg') # サーバーでx環境がない場合などはこれがないとmatplotlibが動かない
import pygrib
import matplotlib.pyplot as plt
import numpy as np
import math
from mpl_toolkits.basemap import Basemap
```

```python
grbs = pygrib.open('/Users/k-ikegami/Desktop/GRIB/Z__C_RJTD_20170120090000_MSM_GPV_Rjp_L-pall_FH00-15_grib2.bin')
```

```python
grbs(name = "Relative humidity")
```

```python
for grb in grbs:
    print(grb)
```

```python
# forecastTimeを一つ選ぶと、物理量ごとに値が得られる。例えばindex=0は海面更生気圧
# その物理量ごとに各地点での値が入っている
grb = grbs.select(forecastTime=0)[0]
```

```python
lats, lons = grb.latlons()
```

```python
# grb["values"]でも同じ
mslp = grb.values
```

```python
mslp
```

```python
# 先の各地点に対して値が得られている
mslp.size
```

```python
# draw
levels = range(math.floor(mslp.min()/100)*100, math.ceil(mslp.max()/100)*100+1,200)

# lat,lonを一次元に変換
# 何故か二次元のままでは描画できない
flat_lats= np.ravel(lats)
flat_lons= np.ravel(lons)

fig = plt.figure()

# マップを作成。4つのパラメータは描画する範囲の指定
m = Basemap(llcrnrlat=lats.min(),urcrnrlat=lats.max(), llcrnrlon=lons.min(),urcrnrlon=lons.max())

# 等圧線をひく
m.contour(flat_lons, flat_lats,mslp,latlon=True,tri=True,levels=levels)

# 海岸線をひく
m.drawcoastlines() 

fig.savefig('/Users/k-ikegami/Desktop/GRIB/mslp.png')
```
